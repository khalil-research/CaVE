```python
import numpy as np
from matplotlib import pyplot as plt
import tol_colors as tc

def plotLoss(prob_name, poly_deg):
    # methods
    methods = ["2s", "cave+","spo+", "pfyl", "nce"]
    labels = ["2-Stage", "CaVE+", "SPO+", "PFYL", "NCE"]
    # color map
    cset =  tc.tol_cset("muted")
    # set fig
    fig = plt.figure(figsize=(16,8))
    # relaxation
    for i, method in enumerate(methods):
        # get path
        res_dir = "./res/{}/n1000deg{}".format(prob_name, poly_deg)
        res_path = res_dir + "/{}_regret_0.csv".format(method)
        # load regret
        regret = np.genfromtxt(res_path, delimiter=",", skip_header=1)
        log_regret = np.log(regret)
        # plot
        plt.plot(log_regret, color=cset[i], alpha=0.8, lw=5, label=labels[i])
    # font size
    plt.xlabel("Epochs", fontsize=38)
    plt.ylabel("Log Normalized Regret", fontsize=38)
    plt.xticks(range(0, 20, 2), fontsize=28)
    plt.yticks(fontsize=24)
    plt.legend(fontsize=28, loc=1)
    # title
    plt.title("Validation Regret for the {} \n with Polynomial Degree {}".format(prob_name.upper(), poly_deg), fontsize=40)
    # save
    plt.tight_layout()
    plt.savefig(res_dir+"/regret_deg{}.pdf".format(poly_deg))
```

## SP5 

### Poly-Degree 4

```python
plotLoss(prob_name="sp5", poly_deg=4)
```

### Poly-Degree 6

```python
plotLoss(prob_name="sp5", poly_deg=6)
```
